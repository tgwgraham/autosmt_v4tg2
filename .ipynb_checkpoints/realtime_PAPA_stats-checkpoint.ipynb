```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from glob import glob
import toml
```

```python
def get_gv_framenum(config):
    """get_gv_framenum(config)
    
    returns arrays containing the frame indices of frames before and after green and violet excitation
    (PAPA and DR, respectively)"""
    
    def add_fw_and_flatten(x,ncycles,fwrange):
        rv = x.reshape(ncycles,1) + fwrange
        rv = rv.flatten()
        return rv
    
    IS = config['illumination_sequence']
    ncycles = IS['ncycles']
    r = IS['r']
    v = IS['v']
    g = IS['g']
    fw = IS['framewindow']
    gfirst = IS['gfirst']
    
    t = 4*r + v + g
    
    fwrange = np.arange(0,fw)
    
    if gfirst: # if the green pulse occurs first in the cycle
        gpre = np.arange(0,ncycles)*t # indices of frames in frame window before green
        gpost = np.arange(0,ncycles)*t + r + g # indices of frames in frame window after green
        vpre = np.arange(0,ncycles)*t + 2*r + g # indices of frames in frame window before violet
        vpost = np.arange(0,ncycles)*t + 3*r + v + g # indices of frames in frame window after violet
    else: # if the violet pulse occurs first in the cycle
        vpre = np.arange(0,ncycles)*t # indices of frames in frame window before violet
        vpost = np.arange(0,ncycles)*t + r + v # indices of frames in frame window after violet
        gpre = np.arange(0,ncycles)*t + 2*r + v # indices of frames in frame window before green
        gpost = np.arange(0,ncycles)*t + 3*r + v + g # indices of frames in frame window after green
    
    gpre = add_fw_and_flatten(gpre,ncycles,fwrange)
    gpost = add_fw_and_flatten(gpost,ncycles,fwrange)
    vpre = add_fw_and_flatten(vpre,ncycles,fwrange)
    vpost = add_fw_and_flatten(vpost,ncycles,fwrange)
    
    return({'vpre':vpre,'vpost':vpost,'gpre':gpre,'gpost':gpost})

def getgv(filename):
    # Get GV ratio and change in localization number due to green and violet pulses
    df = pd.read_csv(filename)
    deltav = df['frame'].isin(fnum['vpost']).sum() - df['frame'].isin(fnum['vpre']).sum()
    deltag = df['frame'].isin(fnum['gpost']).sum() - df['frame'].isin(fnum['gpre']).sum()
    return np.array([deltag/deltav, deltag, deltav])
```

```python
# Calculate change in green and violet localizations and GV ratio
config = toml.load('settings.toml')
fnum = get_gv_framenum(config)
maxfnum = len(glob('../tracking/*.csv'))
allgv = [getgv(f'../tracking/{j}.csv') for j in range(1,maxfnum+1)]
allgv = np.array(allgv)
```

```python
# Plot of GV ratio vs. movie number
# Point area corresponds to number of green + violet localizations
# This is useful, for instance, if you want to monitor changes in PAPA in real time after a perturbation

scaling_factor = 100  # sets point size
ylim = [0,4] # scale for y axis

s = np.abs(allgv[:,2]/scaling_factor)

fig,axs=plt.subplots(1,3,figsize=(16,4))
axs[0].scatter(np.arange(maxfnum)+1,allgv[:,0],s=s,color='blue')
print(f"Number of files: {maxfnum}")
print(f"Mean GV ratio: {np.mean(allgv[:,0]):.4f}")
axs[0].set_xlabel('Movie number')
axs[0].set_ylabel('GV ratio (localizations)');
axs[0].set_ylim(ylim)

axs[1].scatter(np.arange(maxfnum)+1,allgv[:,2],color='darkviolet')
print(f"Mean DR localizations: {np.mean(allgv[:,2]):.1f}")
axs[1].set_xlabel('Movie number')
axs[1].set_ylabel('DR localizations');

axs[2].scatter(np.arange(maxfnum)+1,allgv[:,1],color='green')
print(f"Mean PAPA localizations: {np.mean(allgv[:,1]):.1f}")
axs[2].set_xlabel('Movie number')
axs[2].set_ylabel('PAPA localizations');

plt.savefig('GV_realtime.png',dpi=300,bbox_inches='tight')
```

```python
# Plot change in localizations for PAPA vs. DR
plt.plot(allgv[:,2],allgv[:,1],'.b')
plt.xlabel('$\Delta$localizations, violet')
plt.ylabel('$\Delta$localizations, green');
plt.xlim([0,allgv[:,2].max()*1.2])
plt.ylim([0,allgv[:,1].max()*1.2]);
```

```python
# Plot DR vs. PAPA localizations

maxv = np.inf # maximum number of violet localizations to include in the plot
minv = 0 # minimum number of violet localizations to include in the plot

sel = allgv[:,2]<maxv
sel = sel & (allgv[:,2]>minv)
plt.plot(allgv[sel,2],allgv[sel,1],'b.')
print((allgv[sel,1]/allgv[sel,2]).mean())
plt.xlabel('$\Delta$locs, DR (violet)')
plt.ylabel('$\Delta$locs, PAPA (green)');
print(f"Mean GV ratio of selected cells: {np.mean(allgv[:,0]):.4f}")
```

```python
# Plot overall estimated GV ratio vs. number of files collected (to assess convergence)
plt.plot([(allgv[:j,1]/allgv[:j,2]).mean() for j in range(1,maxfnum)]);
plt.xlabel('Number of movies')
plt.ylabel('Cumulative average GV');
```

```python
# Import displacement distributions

minfnum = 1
maxfnum = allgv.shape[0]
umperpx = 0.16
allgdisp = np.zeros((maxfnum-minfnum+1,50))
allvdisp = np.zeros((maxfnum-minfnum+1,50))
for j in range(minfnum,maxfnum+1):
    try:
        gd = np.genfromtxt(f'../displacements/{j}_gdisp.csv')
        vd = np.genfromtxt(f'../displacements/{j}_vdisp.csv')
        gh,gx = np.histogram(gd*umperpx,bins=np.arange(0,1.01,0.02));
        gh = gh/gh.sum()
        vh,vx = np.histogram(vd*umperpx,bins=np.arange(0,1.01,0.02));
        vh = vh/vh.sum()
        allgdisp[j-minfnum,:] = gh
        allvdisp[j-minfnum,:] = vh
    except:
        print(f'Glitch with {j}.')
```

```python
# Plot displacement distribution heatmap

fig,axs=plt.subplots(1,3,figsize=(12,12))
axs[0].imshow(allvdisp,cmap='Purples',clim=[0,.1])  # adjust clim to the appropriate dynamic range
axs[1].imshow(allgdisp,cmap='Greens',clim=[0,.1])
axs[2].imshow(allgdisp-allvdisp,clim=[-.02,.02])
plt.savefig('displacement_heatmap.png',dpi=300,bbox_inches='tight')
```

```python
# plot cumulative displacement distribution heatmap

fig,axs=plt.subplots(1,3,figsize=(12,12))
axs[0].imshow(allvdisp.cumsum(axis=1),cmap='Purples',clim=[0,1])
axs[1].imshow(allgdisp.cumsum(axis=1),cmap='Greens',clim=[0,1])
axs[2].imshow((allgdisp-allvdisp).cumsum(axis=1),clim=[-.3,.3])
plt.savefig('displacement_heatmap.png',dpi=300,bbox_inches='tight')
```

```python
# Plot cell-by-cell displacement histograms

nmovies = allvdisp.shape[0]
alpha = 1-0.05**(1/nmovies) # adjust transparency based on number of movies

fig,axs=plt.subplots(2,1,figsize=(8,6))
bins = np.arange(0,1.01,0.02)
bincenters = (bins[1:]+bins[:-1])/2
for j in range(nmovies):
    axs[0].plot(bincenters,allvdisp[j,:],color='darkviolet',alpha=alpha)
    axs[0].plot(bincenters,allgdisp[j,:],color='green',alpha=alpha)
    axs[1].plot(bincenters,(allgdisp-allvdisp)[j,:],color='blue',alpha=alpha)
axs[0].plot([],[],color='darkviolet',label='DR')
axs[0].plot([],[],color='green',label='PAPA')
axs[0].legend()
axs[1].plot([],[],color='blue',label='PAPA - DR')
axs[1].legend()
axs[1].plot([0,1],[0,0],color='k');
```

```python
# Plot cell-by-cell cumulative displacement histograms

nmovies = allvdisp.shape[0]
alpha = 1-0.05**(1/nmovies) # adjust transparency based on number of movies

fig,axs=plt.subplots(2,1,figsize=(8,6))
bins = np.arange(0,1.01,0.02)
bincenters = (bins[1:]+bins[:-1])/2
for j in range(allvdisp.shape[0]):
    axs[0].plot(bincenters,allvdisp[j,:].cumsum(),color='darkviolet',alpha=alpha)
    axs[0].plot(bincenters,allgdisp[j,:].cumsum(),color='green',alpha=alpha)
    axs[1].plot(bincenters,(allgdisp-allvdisp)[j,:].cumsum(),color='blue',alpha=alpha)
axs[0].plot([],[],color='darkviolet',label='DR')
axs[0].plot([],[],color='green',label='PAPA')
axs[0].legend()
axs[1].plot([],[],color='blue',label='PAPA - DR')
axs[1].legend()
axs[1].plot([0,1],[0,0],color='k');
```

```python
# plot combined displacement histograms

fig,axs=plt.subplots(2,1,figsize=(8,6))
bins = np.arange(0,1.01,0.02)
bincenters = (bins[1:]+bins[:-1])/2
vtot = 0*bincenters
gtot = 0*bincenters
for j in range(allvdisp.shape[0]):
    
    sel = np.isnan(allvdisp[j,:])
    allvdisp[j,sel] = 0
    
    sel = np.isnan(allgdisp[j,:])
    allgdisp[j,sel] = 0
    
    vtot = vtot + allvdisp[j,:]
    gtot = gtot + allgdisp[j,:]

axs[0].plot(bincenters,vtot,color='darkviolet',alpha=1,label='DR')
axs[0].plot(bincenters,gtot,color='green',alpha=1,label='PAPA')
axs[0].legend()
axs[1].plot(bincenters,gtot-vtot,color='blue',alpha=1,label='PAPA - DR')
axs[1].plot([0,1],[0,0],color='k')
axs[1].legend();
```

```python
# plot combined cumulative displacement distributions

fig,axs=plt.subplots(2,1,figsize=(8,6))
bins = np.arange(0,1.01,0.02)
bincenters = (bins[1:]+bins[:-1])/2
vtot = 0*bincenters
gtot = 0*bincenters
for j in range(allvdisp.shape[0]):
    
    sel = np.isnan(allvdisp[j,:])
    allvdisp[j,sel] = 0
    
    sel = np.isnan(allgdisp[j,:])
    allgdisp[j,sel] = 0
    
    vtot = vtot + allvdisp[j,:]
    gtot = gtot + allgdisp[j,:]

axs[0].plot(bincenters,vtot.cumsum(),color='darkviolet',alpha=1,label='DR')
axs[0].plot(bincenters,gtot.cumsum(),color='green',alpha=1,label='PAPA')
axs[0].legend()
axs[1].plot(bincenters,(gtot-vtot).cumsum(),color='blue',alpha=1,label='PAPA - DR')
axs[1].plot([0,1],[0,0],color='k')
axs[1].legend();
```
